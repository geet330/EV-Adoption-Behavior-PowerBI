```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select your numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 2. Define how many columns you want per row
cols_per_row = 3

# 3. Calculate total rows needed (e.g., 7 columns / 3 = 3 rows needed)
total_plots = len(num_cols)
nrows = math.ceil(total_plots / cols_per_row)

# 4. Create the subplot grid
fig, axes = plt.subplots(nrows=nrows, ncols=cols_per_row, figsize=(15, nrows * 4))

# 5. Flatten the axes array to make indexing simple (turns a 2D grid into a 1D list)
axes = axes.flatten()

# 6. Loop through columns and plot
for i, col in enumerate(num_cols):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i], color='teal')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('')  # Cleans up redundant x-labels

# 7. Hide any empty subplots (e.g., if you have 7 plots in a 3x3 grid, hide the last 2)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# 8. Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select your categorical columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns

# 2. Define how many columns you want per row
cols_per_row = 3

# 3. Calculate total rows needed
total_plots = len(cat_cols)
nrows = math.ceil(total_plots / cols_per_row)

# 4. Create the subplot grid
fig, axes = plt.subplots(nrows=nrows, ncols=cols_per_row, figsize=(15, nrows * 5))

# 5. Flatten the axes array
axes = axes.flatten()

# 6. Loop through categorical columns and plot
for i, col in enumerate(cat_cols):
    # Get the order of categories by frequency so the bars look organized
    cat_order = df[col].value_counts().index

    sns.countplot(data=df, x=col, order=cat_order, ax=axes[i], palette='Set2')

    axes[i].set_title(f'Frequency of {col}')
    axes[i].set_xlabel('')  # Cleans up redundant x-labels
    axes[i].set_ylabel('Count')

    # Rotate text labels by 45 degrees so they don't overlap
    axes[i].tick_params(axis='x', labelrotation=45)

# 7. Hide any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# 8. Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
# Outliers check

import math
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select your numerical columns (outliers only exist in numerical data)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 2. Define how many columns you want per row
cols_per_row = 3

# 3. Calculate total rows needed
total_plots = len(num_cols)
nrows = math.ceil(total_plots / cols_per_row)

# 4. Create the subplot grid
fig, axes = plt.subplots(nrows=nrows, ncols=cols_per_row, figsize=(15, nrows * 4))

# 5. Flatten the axes array
axes = axes.flatten()

# 6. Loop through numerical columns and plot boxplots
for i, col in enumerate(num_cols):
    # Setting x=col creates a horizontal boxplot, which is great for scanning outliers
    sns.boxplot(data=df, x=col, ax=axes[i], color='salmon', flierprops={"marker": "x", "markerfacecolor": "red"})

    axes[i].set_title(f'Outliers in {col}')
    axes[i].set_xlabel('')  # Cleans up redundant x-labels

# 7. Hide any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# 8. Adjust layout and show
plt.tight_layout()
plt.show()
```
